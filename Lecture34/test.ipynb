```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('../datasets/drinks.csv')
df
```

```python
cmap= plt.get_cmap('Spectral')
barchart_for_BW= df.groupby('continent')[['beer_servings','wine_servings']].mean().plot(kind='barh',figsize=(10,6),grid=True,title='Mean of Beer and Wine Servings by Continent',colormap=cmap)
for container in barchart_for_BW.containers:
    barchart_for_BW.bar_label(container)
```

```python
colors=np.where(df['continent']=='EU','blue','black')
colors[df['continent']=='AF']= 'red'
colors[df['country']=='Bulgaria']= 'green'
df.plot(x='beer_servings',y='wine_servings',kind='scatter',c=colors)
```

```python
from matplotlib import colormaps
df.plot.scatter(x='beer_servings',y='wine_servings',
                s=df['total_litres_of_pure_alcohol']*5,
                c='total_litres_of_pure_alcohol',
                colormap='plasma')
```
